```python
import pandas as pd #Para modelar os dados
import matplotlib.pyplot as plt #Para graficos
import plotly.express as px
```

```python
tabela = pd.read_csv('/content/drive/MyDrive/municipio.csv', sep = ',', engine='python')
#(pd.set_option('display.max_rows', 1000))
#(pd.set_option('display.max_columns', 1000))
tabela
```

```python
ribeirao_preto = tabela[tabela.id_municipio == 3543402]
ribeirao_preto
```

```python
ribeirao_preto.info()
```

```python
pd.options.display.float_format = "{:.0f}".format
dados = ribeirao_preto.set_index('id_municipio')
dados
```

```python
dados['ano'] = pd.to_datetime(dados['ano'], format='%Y')
dados.info()
```

```python
dados_selec = dados[['ano', 'pib']]
dados_selec
```

```python
final = dados_selec.set_index('ano')

final
```

```python
plt.figure( figsize=(10, 100) )
final.plot(y='pib')
# Titulo
plt.title('PIB/ANO')
# Eixos
plt.xlabel('ANO')
plt.ylabel('PIB (R$)')
```

```python
plt.figure( figsize=(15,10) )

# Titulo
plt.title('PIB/ANO')
# Eixos
plt.xlabel('ANO')
plt.ylabel('VALOR DO PIB R$')

plt.plot( final,color='#203040');
```
